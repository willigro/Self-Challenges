```python
def image_to_bytearr(file):
    with open(file, "rb") as image:
        f = image.read()
        return bytearray(f)
```

```python
def distance(arr1, arr2):
    if(len(arr1) != len(arr2)):
        return 0
    a = 0
    for i in range(len(arr1)):
        a += math.pow(arr1[i] - arr2[i], 2)
    return math.sqrt(a)
```
